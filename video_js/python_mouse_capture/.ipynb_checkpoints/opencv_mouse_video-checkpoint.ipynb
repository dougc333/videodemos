```python
# the cv2.waitKey(10000) keeps the image there for 10 seconds, if this doesn't work 
# then you have to add cv2.waitKey(1) to other sections to get this to work before detectiong the 
# key press
# keyboard q to kill the image display before timeout
# the codec/file extensions arent well documented
# fourcc if mp4v you need video_writer to have file name m4v extension
# fourcc if mjpg you need video_writer to have file name avi extension
# else you get no output

# for images you can use jpg extension. Does conversion for you. Frame as 2d numpy array works

import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

vc = cv2.VideoCapture("/Users/dc/videodemos/video_mouse.mov")
print '---------properties---------------'
#print vc.get(cv2.cv.CV_CAP_PROP_FRAME_COUNT)
print '----------------------------------'

fourcc = cv2.cv.CV_FOURCC('m', 'p', '4', 'v') #mv4 extension
#fourcc = cv2.cv.CV_FOURCC('M','J','P','G') #avi extenxion
video_writer = cv2.VideoWriter("/Users/dc/videodemos/video_js/frame_dir/output.m4v",fourcc, 15.0, (281,500))
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
try:
    os.stat("/Users/dc/videodemos/video_js/frame_dir")
except:
    os.mkdir("/Users/dc/videodemos/video_js/frame_dir")

num=0
while True:
    ret_code,frame = vc.read()
    print frame.shape
    if ret_code:
        print "writing frame", frame
        cv2.imshow("image",frame)
        video_writer.write(frame)
        cv2.waitKey(10000)    
        key = cv2.waitKey(1) & 0XFF
        if key == ord('q'):
            break
        print str(num)
        f = open("/Users/dc/videodemos/video_js/frame_dir/"+str(num)+'.txt', 'w')
        f.write(str(frame))
        try:
            result = cv2.imwrite("/Users/dc/videodemos/video_js/frame_dir/"+str(num)+".jpg",frame)
            print "result:", result
        except:
            print "bad write"
        f.close()
        break
        
#cv2.waitKey(1) not needed webposts wrong
#cv2.waitKey(1) not needed webposts wrong
#cv2.waitKey(1) not needed webposts wrong
vc.release()
video_writer.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
cv2.waitKey(1)
print "done"
```
